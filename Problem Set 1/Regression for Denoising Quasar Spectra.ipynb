### Linear Regression and Weighted Linear Regression

The normal linear regression has cost function as:
$$
J(\theta) = \frac{1}{2} \sum^m_{i=1}(h_\theta(x^{(i)}) - y^{(i)})^2 = \frac{1}{2}(X\theta - y)^T(X\theta - y)
$$

Since we have $\nabla(x^TAx) = 2Ax$, thus
$$
\nabla J(\theta) = X^TX\theta - X^Ty 
$$
hence, $\theta = (X^TX)^{-1}X^Ty$

In the locally weighted linear regression case, we want to add non-negative weight term as an approach for local regression. The cost function would designed to be
$$
J(\theta) = \sum_{i = 1}^m w^{(i)} (\theta^T x^{(i)} - y^{(i)})^2
$$

Comparing with the regular linear regression, we can construct a diagonal matrix $W$ such that 
$$
W_{ii} = \frac{1}{2} w^{(i)}
$$
So that the vectorized cost function for locally weighted linear regression is
$$
J(\theta) =  (X\theta - y)^TW(X\theta - y)
$$

similarly, we have 
$$
\nabla_\theta J(\theta) = 2X^T(W(X\theta - y))
$$
hence $\theta = (X^TWX)^{-1}X^TWy$

We can see that when $W = I$, all data points are equally weighted, which is exactly the normal case.

In order to construct $W$, we surely want the weight is larger if points are close to each other and vice versa. A fair choice for the weights is 
$$
w^{(i)} = \exp(-\frac{(x^{(i)} - x)^2}{2\tau^2})
$$

If x is vector-valued, the weight can be generalized to be $w^{(i)} = \exp(-\frac{(x^{(i)} - x)^T(x^{(i)} - x)}{2\tau^2})$

### Implementation

We can implement the normal linear regression (vectorized) and the locally weighted regression easily:

```python
def optim(X, y, W = None):
    if W is None:
        return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    else:
        return np.linalg.inv(X.T.dot(W).dot(X)).dot(X.T).dot(W).dot(y)
```

```python
#adding intercept terms
x0 = np.ones(df_tv.shape[1])
x1 = wave_lens
X = np.stack([x0, x1]).T # X contain one column of ones and one column of lambdas (features)

y = df_tv.head(1).values.ravel()  # we take the first row of the table as y
#pandas.Series.ravel.ravel() return the flattened underlying data as an ndarray
```

```python
print(X.shape, y.shape)
```

```python
normalTheta = optim(X,y)
```

```python
plt.plot(wave_lens, df_tv.loc[0].values)
plt.plot(wave_lens, X.dot(normalTheta))
```

```python
# We can construct the Weight matrix W by
def constructWeightMatrix(X, x_eval, tau = 5):
    return np.diag(np.exp(- (X[:,1] - x_eval[1])**2/(2* tau **2)  ))
```

```python
preds = []
for k, x_eval in enumerate(X):
    W = constructWeightMatrix(X, x_eval) # for each x(i) or lambda, we need to compute a Weight Matrix
    theta = optim(X, y, W)               # for each x(i), we need to optimize theta correpsonding to the Weight Matrix for that point
    preds.append(theta.dot(x_eval))      # put the result in the preds
```

```python
plt.plot(wave_lens, df_tv.loc[0].values)
plt.plot(wave_lens, preds)
```

Locally weighted linear regression does a new regression fit for every data point $x^{(i)}$ to evaluate, which is quite expensive. (Typical of non-parametric model)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for k, tau in enumerate([1, 10, 100, 1000]):
    ax = axes[k]
    preds = []
    for x_eval in X:
        W = constructWeightMatrix(X, x_eval, tau=tau)
        theta = optim(X, y, W)
        preds.append(theta.dot(x_eval))

    ax.plot(wave_lens, df_tv.loc[0].values)
    ax.plot(wave_lens, preds)
    ax.set_title('tau = {0}'.format(tau))
```

- (c.i) Smooth spectra by the weighted linear regression:

```python
# define a lwr for single data point of observation:
def lwr_single(X,y,x,tau):
    """locally weighted regression for a single observation"""
    W = constructWeightMatrix(X, x, tau = tau)
    theta = optim(X,y,W)
    return theta

# lwr for multiple xs, aka for a row of observations
def lwr(X,y,xs,tau):
    """lwr for multiple xs"""
    thetas = []
    for x in xs:
        theta = lwr_single(X, y, x, tau)
        thetas.append(theta)
    return thetas
```

```python
# smooth all spectrum record
fs_tv = [] #finished smoothed training values
x0 = np.ones(df_tv.shape[1])
x1 = wave_lens
X = np.stack([x0, x1]).T

# 
for k, row in df_tv.iterrows():
    y = row.values
    thetas = lwr(X,y,X,tau = 5)
    prd = [_t.dot(_x) for (_t, _x) in zip(thetas, X)]
    fs_tv.append(prd)
df_fs_tv = pd.DataFrame(fs_tv, columns=df_tv.columns)
```

- c(ii) In this case we will select the neibours based on the distances calculated from the right spectrum, and then calculate the weighted average of left spectrum.
首先我们来划分一下左右的数据

```python
wl_right = wave_lens[wave_lens >= 1300]
wl_left = wave_lens[wave_lens < 1200]

## separate the table:
df_fs_tv_r = df_fs_tv[wl_right]
df_fs_tv_l = df_fs_tv[wl_left]
```

```python
df_fs_tv_r.shape
```

```python
def ker(t):
    return np.max(1 - t, 0)
```

```python
# It's very similar to k-nearest-neighbour algorithm,
# it select the neihbours based on distances calculated from the right spectrum, 
# and then calculate the weighted (based on distances) average of left spectrum
num_neighb = 3 # number of neighbours to consider
errors = []
preds_tv = []
for k, row in df_fs_tv_r.iterrows():
    dists = ((df_fs_tv_r - row) ** 2).sum(axis=1)
    max_d = dists.max()
    neighb_ds = dists.sort_values()[:num_neighb]

    p1 = np.sum([ker(d / max_d) * df_fs_tv_l.loc[idx] for (idx, d) in neighb_ds.iteritems()], axis=0) 
    p2 = np.sum([ker(d / max_d) for (idx, d) in neighb_ds.iteritems()])
    f_left_hat = p1 / p2
    preds_tv.append(f_left_hat)
    error = np.sum((f_left_hat - df_fs_tv_l.loc[k]) ** 2)
    errors.append(error)
```

```python
np.mean(errors)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(14, 12))
axes = axes.ravel()

for k, idx in enumerate([0, 5, 10, 15, 20, 25, 30, 35, 40]):
    ax = axes[k]
    ax.plot(wave_lens, df_fs_tv.loc[idx], label='smoothed')
    ax.plot(wl_left, preds_tv[idx], label='predicted')
    ax.legend()
    ax.set_title('Example {0}'.format(idx + 1))
plt.tight_layout()
```
